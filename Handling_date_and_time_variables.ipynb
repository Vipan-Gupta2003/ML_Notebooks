# Working with Dates

```python
date['date'] = pd.to_datetime(date['date'])
```

```python
date.info()
```

1. Extract Year

```python
date['date_year'] = date['date'].dt.year
date.head()
```

2. Extract month

```python
date['date_month_no']=date['date'].dt.month

date.head()
```

if we want to get the name of the month then :

```python
date['date_month_name']=date['date'].dt.month_name()

date.head()
```

Extract Days

```python
date['date_day'] =date['date'].dt.day
date.head()
```

Day of Week

```python
date['date_dow'] = date['date'].dt.dayofweek

date.head()
```

Day of Week Name

```python
date['date_dow'] = date['date'].dt.day_name()

date.head()
```

is Weekend ? -> that particular day is a weekend or not.

```python
date['date_is_weekend'] = np.where(date['date_dow'].isin(['Sunday','Saturday']),1,0)

date.drop(columns = ['product_id','city_id','orders']).head()
```

Extract Week of the year

```python
date['date_week'] = date['date'].dt.isocalendar().week

date.drop(columns = ['product_id','city_id','orders']).head()
```

Extract Quater

```python
date['quarter']  =date['date'].dt.quarter
date.drop(columns = ['product_id','city_id','orders']).head()
```

Extract Semester

```python
date['semester']  = np.where(date['quarter'].isin([1,2]),1,2)

date.drop(columns = ['product_id','city_id','orders']).head()
```

Extract Time Elapsed Between Two Dates

```python
import datetime

today = datetime.datetime.today()

today
```

```python
today - date['date']
```

```python
(today - date['date']).dt.days
```

Month Passed

```python
# np.round((today - date['date']) / np.timedelta64(1,'M'),0)
np.round((today - date['date']) / np.timedelta64(1,'W'),0)
```

k-1 Onehot Encoding

```python
pd.get_dummies(df,columns=['fuel','owner'],drop_first=True)

# we generally use one hot encoding using pandas because, pandas does not remember the order of columns ,
# it might gives different columns positions every time we run it
```

OneHot Encoding using sklearn

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(df.iloc[:,0:4],df.iloc[:,-1],test_size=0.2,random_state=0)
```

```python
x_train.head()
```

```python
x_test.head()
```

```python
y_train.head()
```

```python
y_test.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder()
```

```python
ohe.fit_transform(x_train[['fuel','owner']])
```

```python
# to see the sparse matrix generated
ohe.fit_transform(x_train[['fuel','owner']]).toarray()
```

```python
x_train_new = ohe.fit_transform(x_train[['fuel','owner']])
x_test_new = ohe.transform(x_test[['fuel','owner']])
```

```python
x_train[['brand','km_driven']].values
```

```python
# now we will stack x_train and x_train_new together
# np.hstack((x_train[['fuel','km_driven']].values,x_train_new))# this will give error
np.hstack((x_train[['fuel','km_driven']].values,x_train_new.toarray())).shape
```

If we want we can drop first columns using drop = 'first' // to remove multicolinearity.it will remove the first column

```python
ohe = OneHotEncoder(drop='first')
```

```python
x_train_new = ohe.fit_transform(x_train[['fuel','owner']])
x_test_new = ohe.transform(x_test[['fuel','owner']])
```

```python
x_train_new.toarray()
```

as we can see that the output values are in float but if we want to change the datatype we can.

```python
# ohe = OneHotEncoder(drop='first',sparse =True, dtype=np.int32)// sparse is not working
ohe = OneHotEncoder(drop='first', dtype=np.int32)
```

```python
x_train_new = ohe.fit_transform(x_train[['fuel','owner']]).toarray()
x_test_new = ohe.transform(x_test[['fuel','owner']]).toarray()
```

```python
x_train_new
# now the output values are in integer
```

OneHot Encoding with top categories

```python
df['brand'].value_counts()
```

lets take a threshold =100, the count less than hundred car brand will goes to others

```python
counts = df['brand'].value_counts()
```

```python
df['brand'].nunique()
threshold = 100
```

```python
repl = counts[counts<threshold].index
```

```python
pd.get_dummies(df['brand'].replace(repl,'uncommon'))
```


**How to encode categorical data?**

Categorical data consist String values. Machine learning algorithm needs numerical data instead of string values.So we need to encode string values to numerical data , so we can feed it machine learning algorithm.

there many ways to convert categorical data into numerical data but we gonna study two of them :

1. Ordinal Encoding : works on ordinal data

2. One Hot Encoding  : Works on Nominal data

 Label Encoding is similar to ordinal encoding

Let say we have a dataset having input coloumns say 'x' and output column/target value (categorical) say 'y'. then if we want to encode target value/ columns then we cannot use here ordinal encoding , there is another class for it in scikit-learn -> Label Encoder.

It work similarly as ordinal encoder but it is specially designed for output labels.  

Ordinal encoding is used on input label

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('customer.csv')
```

```python
df.sample(5)
```

when we need to do ordinal, label and nominal encoding. we need seprate these columns and then apply them on the columns

```python
df = df.iloc[:,2:]
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(df.iloc[:,0:2],df.iloc[:,-1],test_size=0.2)
```

```python
x_train
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories=[['Poor','Average','Good'],['School','UG','PG']])
# in this object of ordinal encoder class in which we pass a parameter called categories
# which is going to be a list. In this list we will pass some list->according to the no of
# categorical data eg. if we have two columns , we will pass two list
# -we have to pass  order of the values , if not it will be random
```

```python
oe.fit(x_train)
```

```python
df.isna().sum()
```

```python
x_train = oe.transform(x_train)
x_test = oe.transform(x_test)
```

```python
x_train
```

```python
oe.categories_
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
le.fit(y_train)
```

```python
le.classes_
# here we have no option to decide to which value is given to which
# e.g. No: might get 1 and Yes: might get 0
# or vice-versa
```

```python
y_train  =le.transform(y_train)
y_test = le.transform(y_test)
```
